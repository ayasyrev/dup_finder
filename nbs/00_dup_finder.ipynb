```python
#export
ls = lambda x: list(x.iterdir())
Path.ls = ls
Path.lf = lambda x: [i for i in x.iterdir() if i.is_file()]
Path.ld = lambda x: [i for i in x.iterdir() if i.is_dir()]
```
